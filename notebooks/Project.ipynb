Create monthly sales + profit data

```python
import pandas as pd

df = pd.read_csv("cleaned_superstore.csv")

df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_data = df.groupby(['Year', 'Month']).agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

monthly_data['Month_Index'] = range(1, len(monthly_data) + 1)

monthly_data.head()
```

```python
from sklearn.linear_model import LinearRegression

X = monthly_data[['Month_Index']]

y_sales = monthly_data['Sales']
y_profit = monthly_data['Profit']

sales_model = LinearRegression()
profit_model = LinearRegression()

sales_model.fit(X, y_sales)
profit_model.fit(X, y_profit)
```

```python
future_months = pd.DataFrame({
    'Month_Index': range(len(monthly_data) + 1, len(monthly_data) + 7)
})

future_sales = sales_model.predict(future_months)
future_profit = profit_model.predict(future_months)
```

```python
last_year = monthly_data['Year'].iloc[-1]
last_month = monthly_data['Month'].iloc[-1]

future_dates = pd.date_range(
    start=f"{last_year}-{last_month}-01",
    periods=7,
    freq='MS'
)[1:]

forecast_df = pd.DataFrame({
    'Forecast_Date': future_dates,
    'Forecast_Month': future_dates.month_name(),
    'Predicted_Sales': future_sales,
    'Predicted_Profit': future_profit
})

forecast_df
```

```python
forecast_df.to_csv("sales_forecast.csv", index=False)

from google.colab import files
files.download("sales_forecast.csv")
```
